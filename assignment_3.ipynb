# Selenium

```python
from selenium import webdriver
from chromedriver_autoinstaller import install
install(True)
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait, Select
import time
import pandas as pd

from selenium.common.exceptions import NoSuchElementException
from selenium.common.exceptions import ElementClickInterceptedException
from selenium.common.exceptions import ElementNotInteractableException

import requests, pathlib, os, re
import pandas as pd
pd.options.mode.chained_assignment = None
from bs4 import BeautifulSoup
```

```python
opts = Options()
driver = webdriver.Chrome(options=opts)
driver.get("https://www.fih.hockey/outdoor-rankings/netherlands-women-hockey-rankings-48")
driver.title
```

```python
driver.implicitly_wait(0.5)
```

```python
try:
    # Do not consent to cookies
    proceed_to_webpage_button = driver.find_element(By.CSS_SELECTOR, "button[aria-label='Do not consent']")
    proceed_to_webpage_button.click()
except:
    pass
```

```python
try:
    select_year = driver.find_element(By.XPATH, "//*[@id='4623167a-6fe7-41f2-8150-5bdf74fb783b']/div/div/div[2]/div/section/div[2]/div[1]/div[2]/div[1]")
    select_year.click()
    driver.find_element(By.XPATH, "//*[@id='4623167a-6fe7-41f2-8150-5bdf74fb783b']/div/div/div[2]/div/section/div[2]/div[1]/div[2]/div[2]/div/button[6]").click()
except:
    print(":(")
```

```python
while True:
    try:
        more_button = driver.find_element(By.CSS_SELECTOR, ".nav-right")
        more_button.click()
    except (ElementClickInterceptedException, ElementNotInteractableException):
        more_button = driver.find_element(By.CSS_SELECTOR, ".nav-right")
        driver.execute_script("arguments[0].click();", more_button)
    except NoSuchElementException:
        break
    time.sleep(2)
```

# BeautifulSoup

```python


response = requests.get("https://www.fih.hockey/outdoor-rankings/netherlands-women-hockey-rankings-48")

response

html_string = response.text

document = BeautifulSoup(html_string, "html.parser")

dates_2025 = document.find_all("div", attrs={"class": "table-data date"})
year_2025 = document.find(string="2025")
dates_25 = [date.text + " " + year_2025 for date in dates_2025 if date.text != "Date"]

print(dates_25)

year_2024 = document.find(string="2020")
if year_2024 == "2020":
    dates_2024 = document.find_all("div", attrs={"class": "table-data date"})

print(dates_2024)

d = {"Date": dates_25}
nl_fhwomen_df = pd.DataFrame(d)
nl_fhwomen_df
```
